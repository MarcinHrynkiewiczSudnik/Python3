```python
import pandas as pd
import numpy as np
```

```python
def fetch_financial_data(company='AMZN'):
  import pandas_datareader.data as web
  return web.DataReader(name=company, data_source='stooq')

df = fetch_financial_data('ALE')
df.info()
```

```python
df.head()
```

```python
df.to_csv('allegro.csv')
```

```python
df_nov = df[(df.index.month == 11) & (df.index.year == 2021)]
```

```python
df_nov
```

```python
df_nov.to_csv('allegro_nov.csv')
```
